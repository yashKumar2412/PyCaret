```python
from pycaret.datasets import get_data
dataset = get_data('anomaly')
```

```python
from pycaret.anomaly import *
anom_exp = AnomalyExperiment()
anom_exp.setup(dataset)
```

```python
anom_exp.models()
```

```python
model = anom_exp.create_model('knn')
model
```

```python
anomalies = anom_exp.assign_model(model)
anomalies
```

```python
anom_exp.plot_model(model, plot='tsne')
```

```python
prediction = anom_exp.predict_model(model, dataset)
```

```python
prediction.head()
```

```python
anom_exp.save_model(model, 'Anomaly pipeline')
pipeline = anom_exp.load_model('Anomaly pipeline')
pipeline
```
