```python
import pycaret
from pycaret.datasets import get_data
```

```python
get_data('index')
```

```python
dataset = get_data('house')
target = 'SalePrice'
```

```python
from pycaret.regression import *
regr_exp = RegressionExperiment()
regr_exp.setup(dataset, target=target)
```

```python
regr_exp.models()
```

```python
model = regr_exp.create_model('gbr')
```

```python
regr_exp.plot_model(model)
```

```python
regr_exp.plot_model(model, plot='feature')
```

```python
regr_exp.evaluate_model(model)
```

```python
prediction = regr_exp.predict_model(model)
```

```python
prediction.head()
```

```python
regr_exp.save_model(model, 'House pipeline')
pipeline = regr_exp.load_model('House pipeline')
pipeline
```
