```python
import pycaret
from pycaret.datasets import get_data
```

```python
dataset = get_data('pokemon')
```

```python
dataset = dataset.drop(columns=['Name'])
dataset = dataset.drop(columns=['Type 1'])
dataset = dataset.drop(columns=['Type 2'])
dataset = dataset.drop(columns=['#'])
dataset.head()
```

```python
from pycaret.clustering import *
clust_exp = ClusteringExperiment()
clust_exp.setup(dataset)
```

```python
clust_exp.models()
```

```python
km = clust_exp.create_model('kmeans')
```

```python
kmeans_cluster = clust_exp.assign_model(km)
kmeans_cluster
```

```python
clust_exp.plot_model(km, plot = 'cluster')
```

```python
prediction = clust_exp.predict_model(km, dataset)
prediction.head()
```

```python
clust_exp.save_model(km, 'Pokemon pipeline')
pipeline = clust_exp.load_model('Pokemon pipeline')
pipeline
```
